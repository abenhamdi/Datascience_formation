```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
%matplotlib inline
```

```python
x = np.linspace(0, 10, 300)
rng = np.random.RandomState(0)
y = np.cumsum(rng.randn(300, 3), 0)#300 lignes et 3 colonnes
```

```python
plt.plot(x, y)
plt.legend('ABC', ncol=2, loc='upper right')
```
