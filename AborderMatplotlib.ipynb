```python
import matplotlib.pyplot as plt
```

```python
x = range(1,10)
y = [4,3,2,1,0,1,2,3,4]
```

```python
plt.plot(x, y)
```

```python
p = [10,1,4,3,2]
plt.pie(p)
```

```python
Etiquettes = 'Produit 1', 'Produit 2', 'Produit 3', 'Produit 4', 'Produit 5'
p = [10,1,4,3,2]
plt.pie(p, labels= Etiquettes)
plt.show()
```

```python
plt.pie(p, labels= Etiquettes, autopct='%1.1f%%')
```

```python
e = (0, 0, 0, 0, 0.1)
plt.pie(p, explode=e, labels=Etiquettes, autopct='%1.1f%%')#mettre en exargue une partie du pie à aprrofondir
plt.show()
```
