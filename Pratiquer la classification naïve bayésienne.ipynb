```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_blobs#pour generer des données
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)#100 points, 2features, randoù_state pour génerer des nb aléatoires et cluster_std qui indique l'écart type
plt.scatter(X[:, 0], X[:,1], c=y, s=50, cmap='coolwarm');
plt.show()
```

```python
model = GaussianNB() #hypothese correspondant à chaque groupe suivent une simple distrubution Gausienne
model.fit(X, y);
```

```python
rng = np.random.RandomState(0)
Xnouv = [-4, -16] + [12,20] * rng.rand(3000, 2)
ynouv = model.predict(Xnouv)
```

```python
plt.scatter(X[:, 0], X[:,1], c=y, s=50, cmap='coolwarm');
lim = plt.axis()
plt.scatter(Xnouv[:, 0], Xnouv[:,1], c=ynouv, s=20, cmap='coolwarm', alpha=0.1);
plt.axis(lim)
```
