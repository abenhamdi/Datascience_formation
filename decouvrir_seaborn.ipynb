```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 300)
rng = np.random.RandomState(0)
y = np.cumsum(rng.randn(300, 3), 0)
```

```python
plt.plot(x, y)
plt.legend('ABC', ncol=2, loc='upper right');
```
