```python
import tensorflow as tf 
import matplotlib 
import matplotlib.pyplot as plt
```

```python
mnist= tf.keras.datasets.mnist
(x_a,y_a),(x_test,y_test) = mnist.load_data()
```

```python
x_a=tf.keras.utils.normalize(x_a, axis=1)
x_test=tf.keras.utils.normalize(x_test,axis=1)
```

```python
modèle = tf.keras.models.Sequential()
```

```python
modèle.add(tf.keras.layers.Flatten())#applatissage des couche passage de 254*254 a un vecteur de 1 et 784
```

```python
#ajout de la couche dense et on explique que tous les neurones sont liés entre eux
```

```python
#relu= abréviation de linéaire Rectifie fct non linéaire avec des prop intéressantes afin de facilité l'apprentissagedu réseau de neurone
#2emecouche avec les mm param 
#derniere couche permettant de faire la classification sur nos dites catégories+ softmax permet la classification de plusieurs catégories pour les couches de sortie
```

```python
modèle.add(tf.keras.layers.Dense(256, activation= tf.nn.relu))
modèle.add(tf.keras.layers.Dense(128, activation= tf.nn.relu))          
modèle.add(tf.keras.layers.Dense(10, activation= tf.nn.softmax))
modèle.compile(optimizer='adam', loss='space_categorical_crossentropy', metrics=['accuracy'])
modèle.fit(x_a, y_a, epochs=3)
```

```python
val_perte, val_précision = modèle.evaluate(x_test, y_test)
print("perte=", val_perte)
print("précision", val_précision)
```

```python
modèle.save('lecteur_de_chiffres')
```
