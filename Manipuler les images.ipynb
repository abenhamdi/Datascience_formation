```python
cam= data.camera()
plt.imshow(cam)#avec le module data 
```

```python
cam[:40] = 0 #mets en noir les 40 premieres lignes
plt.imshow(cam)
```

```python
cam.shape
```

```python
ind_l = np.arange(len(cam))#Contient les valeurs allant de 1 à 512 => longuer de cam
ind_l
```

```python
ind_c = 2 * ind_l % len(cam)#garde uniquement les valeurs pairs apartir de ind_l
ind_c
```

```python
cam1 = cam.copy()
cam1[ind_l, ind_c] = 0
```
